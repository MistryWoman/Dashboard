```python
import csv
from wordcloud import WordCloud
import pandas as pd
```

```python
def create_wordcloud(conf_name, df):
    "Given the author df for a given conference creates the wordcloud"
    
    keyword_list = []
    
    for word_list in df['Keywords']:
        split = word_list.split(',')
        for keyword in split:
            keyword = keyword.strip()
            if keyword != "[]":
                keyword_list.append(keyword)
                
    wordcloud_text = " ".join(keyword_list)
    wordcloud = WordCloud(background_color = "white", max_words = 500, max_font_size = 100, 
                      relative_scaling = 0, min_font_size = 15, colormap = 'viridis', width= 1000, height= 800).generate(wordcloud_text)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.figure(figsize = (10000, 10000))
    plt.show()
    file_name = str(conf_name) + '_wordcloud.png'
    wordcloud.to_file("assets/" + str(file_name))
    

ipsn_auth = pd.read_csv('data/Clean_Authors/ipsn_authors.csv')
create_wordcloud('ipsn', ipsn_auth)

mobicom_auth = pd.read_csv('data/Clean_Authors/mobicom_authors.csv')
create_wordcloud('mobicom', mobicom_auth)

mobihoc_auth = pd.read_csv('data/Clean_Authors/mobihoc_authors.csv')
create_wordcloud('mobihoc', mobihoc_auth)

sensys_auth = pd.read_csv('data/Clean_Authors/sensys_authors.csv')
create_wordcloud('sensys', sensys_auth)

sigcomm_auth = pd.read_csv('data/Clean_Authors/sigcomm_authors.csv')
create_wordcloud('sigcomm', sigcomm_auth)




```
